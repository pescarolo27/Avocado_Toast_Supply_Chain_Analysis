```python
#import packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load in the data

#CSV's
avocado_orig = pd.read_csv("avocado.csv", sep="\t")
lemon_orig = pd.read_csv("lemon.csv", delimiter="\t")
olive_oil_orig = pd.read_csv("olive_oil.csv", delimiter="\t")
salt_flakes_orig = pd.read_csv("salt_flakes.csv", delimiter="\t")
sourdough_orig = pd.read_csv("sourdough.csv", delimiter="\t")

#TXT's
avocado_cats = pd.read_csv("relevant_avocado_categories.txt", header=None).rename(columns={0:"category"})
olive_oil_cats = pd.read_csv("relevant_olive_oil_categories.txt", header=None).rename(columns={0:"category"})
sourdough_cats = pd.read_csv("relevant_sourdough_categories.txt", header=None).rename(columns={0:"category"})

display(avocado_orig, avocado_cats)
```

## Cleaning & Filtering
### Avocado
Recall that the original shape is 1785 rows x 184 columns.

```python
#drop null 'categories_tags' observations from avocado data
avocado_I = avocado_orig[avocado_orig.categories_tags.notna()].reset_index(drop=True)
    #display(len(avocado_orig), len(avocado))

#make a new column for 'categories_tags' that is a list of each observation
avocado_I['categories_list'] = avocado_I.categories_tags.apply(lambda x: x.split(","))
#display(avocado_I[['code','categories_tags','categories_list']])

#filter for observations which have 'categories_tags' in the .txt categories
    #for each row, have to search through each element in 'category_list' & see if they match any element in corresponding .txt
avocado_I['in_cats?'] = avocado_I.categories_list.apply(lambda x: any(i in [*avocado_cats.category] for i in x))
#display(avocado_I[['code','categories_list','in_cats?']])

#using this new column, 'in_cats?', filter for rows that have a category tag ('categories_tags'/'categories_list') in 
    #the relevant category tags specified in the corresponding .txt file
avocado = avocado_I[avocado_I['in_cats?'] == True].reset_index(drop=True)
    #are 182 such rows

#subset for main relevant columns, as specified in Objectives
avocado = avocado[['code', 'lc', 'product_name_en', 'quantity', 'serving_size', 'packaging_tags', 'brands', 'brands_tags', 
                   'categories_tags', 'labels_tags', 'countries', 'countries_tags', 'origins', 'origins_tags']]
display(avocado)
```

### Olive Oil
Recall that the original shape is 8,298 rows x 326 columns.

```python
#drop null 'categories_tags' observations from olive-oil data
olive_oil_I = olive_oil_orig[olive_oil_orig.categories_tags.notna()].reset_index(drop=True)
    #display(len(olive_oil_orig), len(olive_oil))

#make a new column for 'categories_tags' that is a list of each observation
olive_oil_I['categories_list'] = olive_oil_I.categories_tags.apply(lambda x: x.split(","))
#display(olive_oil_I[['code','categories_tags','categories_list']])

#filter for observations which have 'categories_tags' in the .txt categories
    #for each row, have to search through each element in 'category_list' & see if they match any element in corresponding .txt
olive_oil_I['in_cats?'] = olive_oil_I.categories_list.apply(lambda x: any(i in [*olive_oil_cats.category] for i in x))
#display(olive_oil_I[['code','categories_list','in_cats?']])

#using this new column, 'in_cats?', filter for rows that have a category tag ('categories_tags'/'categories_list') in 
    #the relevant category tags specified in the corresponding .txt file
olive_oil = olive_oil_I[olive_oil_I['in_cats?'] == True].reset_index(drop=True)
    #are 5,290 such rows

#subset for main relevant columns, as specified in Objectives
olive_oil = olive_oil[['code', 'lc', 'product_name_en', 'quantity', 'serving_size', 'packaging_tags', 'brands', 'brands_tags', 
                   'categories_tags', 'labels_tags', 'countries', 'countries_tags', 'origins', 'origins_tags']]
display(olive_oil)
```

### Sourdough
Recall that the original shape is 1,422 rows x 189 columns.

```python
#drop null 'categories_tags' observations from sourdough data
sourdough_I = sourdough_orig[sourdough_orig.categories_tags.notna()].reset_index(drop=True)
    #display(len(sourdough_orig), len(sourdough))

#make a new column for 'categories_tags' that is a list of each observation
sourdough_I['categories_list'] = sourdough_I.categories_tags.apply(lambda x: x.split(","))
#display(sourdough_I[['code','categories_tags','categories_list']])

#filter for observations which have 'categories_tags' in the .txt categories
    #for each row, have to search through each element in 'category_list' & see if they match any element in corresponding .txt
sourdough_I['in_cats?'] = sourdough_I.categories_list.apply(lambda x: any(i in [*sourdough_cats.category] for i in x))
#display(sourdough_I[['code','categories_list','in_cats?']])

#using this new column, 'in_cats?', filter for rows that have a category tag ('categories_tags'/'categories_list') in 
    #the relevant category tags specified in the corresponding .txt file
sourdough = sourdough_I[sourdough_I['in_cats?'] == True].reset_index(drop=True)
    #are 399 such rows

#subset for main relevant columns, as specified in Objectives
sourdough = sourdough[['code', 'lc', 'product_name_en', 'quantity', 'serving_size', 'packaging_tags', 'brands', 'brands_tags', 
                   'categories_tags', 'labels_tags', 'countries', 'countries_tags', 'origins', 'origins_tags']]
display(sourdough)
```

## Analysis
*Recall:* Apply your data manipulation and analysis skills to investigate the supply chain of ingredients for making avocado toast in the U.K. Your task is to determine the following information:
- The name of the most common country of origin for each of the three key ingredients: avocados, olive oil, and sourdough.

Store the most common country of origin for each ingredient in the following variables: `top_avocado_origin`, `top_olive_oil_origin`, and `top_sourdough_origin`. Ensure that the country names contain only letters (A-Z) and spaces, with no hyphens or other characters.

**Notes:**
- Recall that the `origins_tags` variable refers to the country of origin for each item.
- Before counting, want to filter for observations from the UK. Can use the `countries` variable to do so.

```python
## AVOCADO - after filtering & cleaning, has 182 rows
display("AVOCADO:")
display(avocado[avocado.countries == "United Kingdom"].origins_tags.value_counts(dropna=False))
display("--------------------------------------------------------------")

#OLIVE OIL - after filtering & cleaning, has 5,290 rows
display("OLIVE OIL:")
display(olive_oil[olive_oil.countries == "United Kingdom"].origins_tags.value_counts(dropna=False))
display("--------------------------------------------------------------")

#SOURDOUGH - after filtering & cleaning, has 399 rows
display("SOURDOUGH:")
display(sourdough[sourdough.countries == "United Kingdom"].origins_tags.value_counts(dropna=False))
```

```python
#assign the specified variables
top_avocado_origin = avocado[avocado.countries == "United Kingdom"].origins_tags.value_counts(
    dropna=True).index[0].split(":")[1].title()
top_olive_oil_origin = olive_oil[olive_oil.countries == "United Kingdom"].origins_tags.value_counts(
    dropna=True).index[0].split(":")[1].title()
top_sourdough_origin = sourdough[sourdough.countries == "United Kingdom"].origins_tags.value_counts(
    dropna=True).index[0].split(":")[1].replace("-", " ").title()

display("Most common country of origin for avocados: {}.".format(top_avocado_origin))
display("Most common country of origin for olive oil: {}.".format(top_olive_oil_origin))
display("Most common country of origin for sourdough: {}.".format(top_sourdough_origin))
```
